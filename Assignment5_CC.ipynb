```python
import numpy as np
array = np.array([10, 52, 62, 16, 16, 54, 453])
print("Sorted array : ",np.sort(array))
indices = np.argsort(array)
print(indices)
k=4
result = np.argpartition(array,k)
print("Smallest values:",array[result[:k]])
n=5
values = np.sort(array)[::-1][:n]
print("Largest Values :",values)
```

```python
import numpy as np
first_name = "Numpy"
last_name = "Python"
X = sum(ord(c) for c in first_name[0] + last_name[0])
sales = np.array([X, X+50 ,X+100, X+150, X+200])
print("Data of initial sales:",sales)

tax_rate = ((X % 5) + 5)/100
sales_with_tax = sales*(1 + tax_rate)

print("Sales with tax:",sales_with_tax)

discount_5 = 0.05
discount_10 = 0.10

discounted_sales = np.where(sales < X +100, sales_with_tax * (1 - discount_5), sales_with_tax*(1- discount_10))
print("Sales after discount:",discounted_sales)

num_weeks = 3
weekly_sales = np.tile(discounted_sales, (num_weeks, 1))  

weekly_increase = np.array([1.0, 1.02, 1.04]).reshape(-1, 1)
weekly_sales_increased = weekly_sales * weekly_increase

print("Sales for 3 Weeks:", weekly_sales_increased)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

def y_x_squared(x):
    return x**2

def y_sin_x(x):
    return np.sin(x)

def y_exp_x(x):
    return np.exp(x)

def y_log_x(x):
    return np.log(np.abs(x) + 1) 

def plot_function(x, y, title, xlabel="x", ylabel="y"):
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()
y1 = y_x_squared(x)
plot_function(x, y1, "Y = x^2")

y2 = y_sin_x(x)
plot_function(x, y2, "Y = sin(x)")

y3 = y_exp_x(x)
plot_function(x, y3, "Y = e^x")

y4 = y_log_x(x)
plot_function(x, y4, "Y = log(|x| + 1)")
```
